```python
import matplotlib
from matplotlib import style
style.use('seaborn')
import matplotlib.pyplot as plt
import pandas as pd 
from sqlalchemy import create_engine 
```

```python
airlines_file = "airlines.csv"
airlines_df = pd.read_csv(airlines_file)
```

```python
#Airline clean
# Create a filtered dataframe from specific columns
airlines_cols = ['Airline ID', 'Name', 'ICAO', 'IATA', 'Country', 'Active']
airlines_trans= airlines_df[airlines_cols].copy()

# Rename the column headers
airlines_trans = airlines_trans.rename(columns={"Airline ID": "airline_id",
                                            "Name": "airline_name",
                                            "ICAO":"icao",
                                            "IATA":"iata",
                                            "Country":"country",
                                            "Active":"active"
                                             })

# Clean the data by dropping duplicates and setting the index
airlines_trans.drop_duplicates("airline_id", inplace=True)
airlines_trans.set_index("airline_id", inplace=True)
airlines_trans=airlines_trans.dropna()
#drop all the airlines that country is diferent than United States and drop the column country
airlines_trans = airlines_trans.loc[airlines_trans["country"]=="United States", :]
airlines_trans=airlines_trans.drop(['country'], axis=1)
airlines_trans.head(10)
```
